```python
import pandas as pd
import statsmodels.formula.api as smf

df_panel = pd.read_csv('final_city_year_panel.csv')

df_panel['century'] = df_panel['year'] // 100 + 1
```

```python
df_subset = df_panel[(df_panel['century'].between(13, 19)) & (df_panel['citypop_le10'] > 0)]
```

```python
control_vars = ['C(century)', 'latitude', 'longitude']
```

```python
import statsmodels.formula.api as smf


group_vars = ['university', 'citypop_le10']
control_vars = ['C(century)', 'latitude', 'longitude']


formula1= 'creative_people_count ~ ' + ' + '.join(control_vars + group_vars)


model_edu = smf.ols(formula=formula1, data=df_subset).fit()


print(model_edu.summary())

#'university', 'citypop_le10' both keep
```

```python

formula2 = '''
creative_people_count ~ commune + free_prince_dls + parl_act + C(century) + latitude + longitude
'''

model_institution = smf.ols(formula=formula2, data=df_subset).fit()

print(model_institution.summary())
```

```python

formula21 = '''
creative_people_count ~ commune + free_prince_dls + C(century) + latitude + longitude
'''

model_institution = smf.ols(formula=formula21, data=df_subset).fit()

print(model_institution.summary())

#only keep 'commune','free_prince_dls', 
# delete 'parl_act'
```

```python

religious_vars = ['bishop', 'archbishop', 'muslim_holy_city']


formula_religion = 'creative_people_count ~ ' + ' + '.join(religious_vars + control_vars)


model_religion = smf.ols(formula=formula_religion, data=df_subset).fit()


print(model_religion.summary())

#only keep 'archbishop'
```

```python

transport_vars = ['river', 'caravan_hub', 'rom_road_nohub']



formula_transport = 'creative_people_count ~ ' + ' + '.join(transport_vars + control_vars)


model_transport = smf.ols(formula=formula_transport, data=df_subset).fit()


print(model_transport.summary())

# keep 'river', 'caravan_hub' for now
#'caravan_hub' is not sure
```

```python

market_vars = ['fup', 'chrfup_devries', 'musfup_devries']


formula_market = 'creative_people_count ~ ' + ' + '.join(market_vars + control_vars)


model_market = smf.ols(formula=formula_market, data=df_subset).fit()


print(model_market.summary())

# keep 'chrfup_devries'
```

```python



agglomeration_vars = [
    'nrcities_le10_0_20',
    'citypop_le10_0_20',
    'size_nearest_le10_mus',
    'size_nearest_le10_chr'
]

formula_agglomeration = 'creative_people_count ~ ' + ' + '.join(agglomeration_vars + control_vars)


model_agglomeration = smf.ols(formula=formula_agglomeration, data=df_subset).fit()


print(model_agglomeration.summary())

# all drops
```

```python

cultural_vars = ['capital','plundered'] 


formula_culture = 'creative_people_count ~ ' + ' + '.join(cultural_vars + control_vars)


model_culture = smf.ols(formula=formula_culture, data=df_subset).fit()


print(model_culture.summary())

#keep 'capital'
```

```python
#final model
final_formula = 'creative_people_count ~ university + citypop_le10 + commune + free_prince_dls + archbishop + chrfup_devries + river + caravan_hub + capital + latitude + C(century)'


final_model = smf.ols(formula=final_formula, data=df_subset).fit()


print(final_model.summary())
```

```python
#final model

final_formula2 = 'creative_people_count ~ university + citypop_le10 + commune + free_prince_dls + archbishop +  river + caravan_hub + capital + latitude + C(century)'

# OLS regression
final_model = smf.ols(formula=final_formula2, data=df_subset).fit()


print(final_model.summary())
```

```python

# make sure the path accumulation variable has been generated
df_subset.loc[:, 'previous_century_creative_count'] = df_subset.groupby('city')['creative_people_count'].shift(1)
df_subset.loc[:, 'previous_century_creative_count'] = df_subset['previous_century_creative_count'].fillna(0)

# zoom pop/1000
df_subset.loc[:, 'previous_century_creative_count_scaled'] = df_subset['previous_century_creative_count'] / 1000
df_subset.loc[:, 'citypop_le10_scaled'] = df_subset['citypop_le10'] / 1000

# regression formula
formula = '''
creative_people_count ~ university + commune + free_prince_dls + citypop_le10_scaled 
+ previous_century_creative_count_scaled + caravan_hub + archbishop + capital + latitude + C(century)
'''

# ols model
model = smf.ols(formula=formula, data=df_subset).fit()


print(model.summary())
```

```python
import numpy as np

# Add 1 to avoid log(0)
df_subset['log_creative_count'] = np.log(df_subset['creative_people_count'] + 1)
df_subset['log_prev_creative_count'] = np.log(df_subset['previous_century_creative_count'] + 1)

# zoom pop/1000
df_subset['citypop_le10_scaled'] = df_subset['citypop_le10'] / 1000

# Regression formula (log-log path accumulation expression)
formula_log = '''
log_creative_count ~ university + commune + free_prince_dls + citypop_le10_scaled 
+ log_prev_creative_count + caravan_hub + archbishop + capital + latitude + C(century)
'''

# ols
model_log = smf.ols(formula=formula_log, data=df_subset).fit()


print(model_log.summary())
```

```python

#For verification
# create a Top20 Dummy
# Rule: Is the creative_people_count in the last century in the top 20? 
df_subset['top20_dummy'] = df_subset.groupby('century')['previous_century_creative_count']\
    .transform(lambda x: x >= x.sort_values(ascending=False).iloc[19]).astype(int)


formula_dummy = '''
creative_people_count ~ university + commune + free_prince_dls + citypop_le10_scaled 
+ top20_dummy + caravan_hub + archbishop + capital + latitude + C(century)
'''


model_dummy = smf.ols(formula=formula_dummy, data=df_subset).fit()


print(model_dummy.summary())

#dependent variable -rank top20
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from patsy import dmatrices


y, X = dmatrices(formula_log, data=df_subset, return_type='dataframe')

# caculate VIF of each variavble
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```

```python
df_ranked = df_subset.copy()

df_ranked['rank_within_century'] = df_ranked.groupby('century')['creative_people_count'] \
                                             .rank(method='min', ascending=False)
```

```python

df_ranked['rank_inverse'] = 1 / df_ranked['rank_within_century']


formula_rank = '''
rank_within_century ~ university + commune + free_prince_dls + citypop_le10_scaled 
+ log_prev_creative_count + caravan_hub + archbishop + capital + latitude + C(century)
'''

model_rank = smf.ols(formula=formula_rank, data=df_ranked).fit()
print(model_rank.summary())
```
